## Split Dataset using StratifiedShuffleSplit (index-based)

```python
from sklearn.model_selection import StratifiedShuffleSplit
sss1 = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=42)
for train_idx, temp_idx in sss1.split(data, data[TARGET_COL]):
    train_df = data.iloc[train_idx]
    temp_df = data.iloc[temp_idx]

sss2 = StratifiedShuffleSplit(n_splits=1, test_size=0.5, random_state=42)
for val_idx, test_idx in sss2.split(temp_df, temp_df[TARGET_COL]):
    val_df = temp_df.iloc[val_idx]
    test_df = temp_df.iloc[test_idx]

os.makedirs(sssPath, exist_ok=True)
train_df.to_csv(f"{sssPath}/train.csv", index=False)
val_df.to_csv(f"{sssPath}/val.csv", index=False)
test_df.to_csv(f"{sssPath}/test.csv", index=False)
```

## Split Dataset manually

```python
PATH = '../dataset/Dataset_TehHijau.csv'
TARGET_COL = "Kategori"

import pandas as pd
data = pd.read_csv(PATH)

import numpy as np
import os

commonPath = "../dataset"
random_state = [42]

for state in random_state:
  manualPpath = f"{commonPath}/manual/{state}"
  # Create shuffled indices
  indices = np.arange(len(data))
  rng = np.random.default_rng(seed=state)
  rng.shuffle(indices)

  n = len(indices)

  # Compute split points
  train_end = int(0.7 * n)
  val_end   = int(0.85 * n)

  # Split indices
  train_idx = indices[:train_end]
  val_idx   = indices[train_end:val_end]
  test_idx  = indices[val_end:]

  print("Train idx:", train_idx)
  print("Val idx:", val_idx)
  print("Test idx:", test_idx)

  # Create datasets
  train_df = data.iloc[train_idx]
  val_df   = data.iloc[val_idx]
  test_df  = data.iloc[test_idx]

  print(f"Train: {len(train_df)} rows")
  print(f"Val:   {len(val_df)} rows")
  print(f"Test:  {len(test_df)} rows")

  os.makedirs(manualPpath, exist_ok=True)
  train_df.to_csv(f"{manualPpath}/train.csv", index=False)
  val_df.to_csv(f"{manualPpath}/val.csv",     index=False)
  test_df.to_csv(f"{manualPpath}/test.csv",   index=False)
```
